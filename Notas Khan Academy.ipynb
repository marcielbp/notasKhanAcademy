```python
import pandas as pd
import numpy as np
notas = pd.read_csv("Baixados.csv",sep=",",low_memory=False)
discentes = pd.read_csv("discentes.csv",sep=",",low_memory=False)
dF = pd.DataFrame(notas)
dF = dF.fillna(0)
dF2 = dF[dF !=999]
```

```python

#Recomendações parte 1 - Pré Cálculo
for i in range (0,len(discentes)):
    a = discentes.iloc[i]['NOME']
    notasDF = dF[dF['Nome do aluno'].str.contains(a).fillna(False)]
    #notasDF = notasDF[notasDF.iloc[:,7].str.contains('Ago')]
    notasDF = notasDF[notasDF.iloc[:,7].str.contains('Nov')]
    #notasDF1 = notasDF[notasDF.iloc[:,7].str.contains('Set')]
    #notasDF2 = notasDF[notasDF.iloc[:,7].str.contains('Out')] 
    #notasDF = notasDF1.append(notasDF2)
    numRec = len(notasDF)
    recDone = notasDF[notasDF.iloc[:,6]!=0]
    numDone = len(recDone)
    recTeste = recDone[recDone.iloc[:,9].str.contains('Teste')]
    recExe = recDone[recDone.iloc[:,9].str.contains('Exerc')]
    recTot = recTeste.append(recExe)
    if (numRec <= 0.0):
        propREC = 0.0 
    else:
        propREC = numDone/numRec
    
    totREC = sum(recTot.iloc[:,4])
    if (totREC <= 0.0):
        propEXE = 0.0
    else:
        propEXE = sum(recTot.iloc[:,3]/recTot.iloc[:,4])/len(recTot)
    if (propEXE>=0.75):
        propEXE=1.0
    #print(a)
    #print('Percentual das recomendações',100*propREC,'%')
    #print('Percentual dos testes/ Exercícios',100*propEXE,'%')
    nota = max(propREC,propEXE)
    #print(a,'\t {0:1.1f}'.format(10*nota))
    print('{0:1.1f}'.format(10*nota))
```
